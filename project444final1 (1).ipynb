```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
import pandas as pd


df_sorted = df.sort_values(by=['Road_Accidents__during_2018'], ascending=False)[1:21]

plt.figure(figsize=(12, 6))

plt.bar(df_sorted["State"], df_sorted["Road_Accidents__during_2018"], color='blue')
plt.xlabel("States")
plt.ylabel("Number of Accidents")
plt.title("Road Accidents in 2018 (Sorted)")
plt.xticks(rotation=90)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(3, 1, figsize=(10, 13))
sns.barplot(data=low_2018, x='State', y='Road_Accidents__during_2018', ax=axes[0])
axes[0].set_title("Lowest Accident in 2018 State Wise", fontsize=14, pad=14)
axes[0].set_xlabel('State', fontsize=12)
axes[0].set_ylabel('Accidents During 2018', fontsize=12)
sns.barplot(data=low_2019, x='State', y='Road_Accidents__during_2019', ax=axes[1])
axes[1].set_title("Lowest Accident in 2019 State Wise", fontsize=14, pad=14)
axes[1].set_xlabel('State', fontsize=12)
axes[1].set_ylabel('Accidents During 2019', fontsize=12)
sns.barplot(data=low_2020, x='State', y='Road_Accidents__during_2020', ax=axes[2])
axes[2].set_title("Lowest Accident in 2020 State Wise", fontsize=14, pad=14)
axes[2].set_xlabel('State', fontsize=12)
axes[2].set_ylabel('Accidents During 2020', fontsize=12)
for ax in axes:
    ax.tick_params(axis='x', which='major', labelsize=12, rotation=90)
    ax.tick_params(axis='y', which='major', labelsize=12)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.tight_layout()
fig, axes = plt.subplots(3, 1, figsize=(10, 10), sharex=True)
axes[0].plot(df["Total_Number_of_Accidents__Per_Lakh_Population_-_2018"], marker='o', color='blue', linestyle='-', label="2018")
axes[0].set_ylabel("Accidents per Lakh")
axes[0].set_title("Total Accidents per Lakh Population - 2018")
axes[0].set_xlabel("State")

axes[1].plot(df["State"], df["Total_Number_of_Accidents__Per_Lakh_Population_-_2019"], marker='s', color='green', linestyle='-', label="2019")
axes[1].set_ylabel("Accidents per Lakh")
axes[1].set_title("Total Accidents per Lakh Population - 2019")
axes[1].set_xlabel("State")

axes[2].plot(df["State"], df["Total_Number_of_Accidents__Per_Lakh_Population_-_2020"], marker='d', color='red', linestyle='-', label="2020")
axes[2].set_ylabel("Accidents per Lakh")
axes[2].set_title("Total Accidents per Lakh Population - 2020")
axes[2].set_xlabel("State")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
df_sorted_2018 = df.sort_values(by=['Share_in_Death-_2018'], ascending=False)[1:21]
df_sorted_2019 = df.sort_values(by=['Share_in_Death-_2019'], ascending=False)[1:21]
df_sorted_2020 = df.sort_values(by=['Share_in_Death-_2020'], ascending=False)[1:21]

bar_width = 0.25
index = np.arange(len(df_sorted_2018["State"]))

plt.figure(figsize=(12, 6))
plt.bar(index - bar_width, df_sorted_2018["Share_in_Death-_2018"], bar_width, color='blue', label="2018")
plt.bar(index, df_sorted_2019["Share_in_Death-_2019"], bar_width, color='green', label="2019")
plt.bar(index + bar_width, df_sorted_2020["Share_in_Death-_2020"], bar_width, color='red', label="2020")

plt.xlabel("States")
plt.ylabel("Share in Deaths")
plt.title("Death Share Comparison (2018-2020) ")
plt.xticks(index, df_sorted_2018["State"], rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
total_accidents_2018 = df['Road_Accidents__during_2018'].sum()
total_accidents_2019 = df['Road_Accidents__during_2019'].sum()
total_accidents_2020 = df['Road_Accidents__during_2020'].sum()
years = ['2018', '2019', '2020']
accidents = [total_accidents_2018, total_accidents_2019, total_accidents_2020]

plt.figure(figsize=(6, 6))
plt.pie(accidents, labels=years, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Road Accidents - By Year')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

total_share_2018 = df['Share_in_Death-_2018'].sum()
total_share_2019 = df['Share_in_Death-_2019'].sum()
total_share_2020 = df['Share_in_Death-_2020'].sum()


years = ['2018', '2019', '2020']
shares = [total_share_2018, total_share_2019, total_share_2020]

plt.figure(figsize=(6, 6))
plt.pie(shares, labels=years, autopct='%1.1f%%', startangle=90)
plt.title('Share of Deaths in Road Accidents - By Year')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df['Total_Deaths'] = df['Share_in_Death-_2018'] + df['Share_in_Death-_2019'] + df['Share_in_Death-_2020']

df_sorted = df.sort_values(by=['Total_Deaths'], ascending=False)

top_10_states = df_sorted.head(10)

states = top_10_states['State']
deaths = top_10_states['Total_Deaths']

plt.figure(figsize=(8, 8))
plt.pie(deaths, labels=states, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Deaths in Road Accidents Across States (2018-2020)')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

heatmap_columns = [
    'Road_Accidents__during_2018', 'Road_Accidents__during_2019', 'Road_Accidents__during_2020',
    'Persons_Killed_2018', 'Persons_Killed_2019', 'Persons_Killed_2020',

]

correlation_matrix = df[heatmap_columns].corr()

print("Correlation Analysis:\n", correlation_matrix)

plt.figure(figsize=(8, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Road Accident Data')
plt.show()

```
